From the dataset, it is implied that patient with the diabetes pedigree function of  average mean value of 0.55 have the great risk of being diabetic, and also an individual with less than 0.088 pedigree function value is at reduced diabetes risk. The 50% percentile rank of pedigree function that results in positive diabetic status is between the value range 0.2625 - 0.72800.  The pedigree function values is a stable variable that can be used to dtermine  a diabetic status,but the real-valued attribute tends discrete,  as result of the fact that it will take much period of time for a family diabetic history to be altered. For instance,  it is convetionaly rare to identify an individual with a low valued pedigree function (<0.088) to test positive to diabetes. Hence, the diabetic  posistive status of patient  cannot be solely determined by the pedigree function, due to the fact a discplined individual can disrupt the trend. The next ranked high unique valued variables are; Body mass Index,Insulin, and Glucose.  (Bays et.al,2007) has argued that the increased BMI was associated with increased prevalence of diabetes, while the glucose variable  measures the  blood sugar level for diabetic testing. Considering the distribution of the glucose feature of average mean value of 141.257, with high standard deviation of 31.93, and the insulin parameter has average mean value of 79.8, with standard deviation values of 115.24. The average mean  values of both variables seem to be less  robust and unreliable to determining a accurate diabetic status. Conversely, in the relationship analysis of all the variables  the glucose variable the is the most correlated feature in respect to the diabetic outcome at the statistical significance of 0.47. 

Hence, this study adopts the BMI with the diabetic test outcome to perform the realtionship analysis in respect to other variables.

```python
def create_model(lrn_rate,input_layer):
    model = tf.keras.models.Sequential()
    model.add(input_layer)
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden1'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden2'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden3'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden4'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden5'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden6'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden7'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden8'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden9'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden10'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden11'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden12'))
    model.add(tf.keras.layers.Dense(units=36,activation='relu',name='Hidden13'))
    model.add(tf.keras.layers.Dense(units=1,activation='softmax',name='ouput'))
    model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate = lrn_rate),loss=tf.keras.losses.BinaryCrossentropy(), 
                  metrics =[Accuracy(name='acc')])
    return model
```
